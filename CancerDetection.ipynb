```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data = load_breast_cancer()
data
```

```python
df = pd.DataFrame(np.c_[data.data, data.target], columns = [list(data.feature_names)+['target']])
df
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(df.corr(), annot = True)
```

```python
X = df.drop(columns = ['target'])
y = df.iloc[:, -1]
```

## Split and Train Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 9)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Scaling is not necessary in decison Tree and random forest classification
from sklearn.preprocessing import MinMaxScaler
mmc = MinMaxScaler()
mmc.fit(X_train)
X_train = mmc.transform(X_train)
X_test = mmc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
accuracy_score(y_test,y_pred_lr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = lr.predict(X_test)
accuracy_score(y_test,y_pred_knn)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'gini')
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_score(y_test,y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion = 'gini', n_estimators = 100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_score(y_test,y_pred_rf)
```

```python
from sklearn.svm import SVC
```

```python
# kernal is used to specify type of algo

svc_r = SVC(kernel='rbf')
svc_r.fit(X_train, y_train)
y_pred_svc_r = svc_r.predict(X_test)
accuracy_score(y_test, y_pred_svc_r)
```

```python
svc_l = SVC(kernel='linear')
svc_l.fit(X_train, y_train)
y_pred_svc_l = svc_l.predict(X_test)
accuracy_score(y_test, y_pred_svc_l)
```

```python
# This can also be used with polynomial.
# kernal is used to specify type of algo

svc_p = SVC(kernel='poly')
svc_p.fit(X_train, y_train)
y_pred_svc_p = svc_p.predict(X_test)
accuracy_score(y_test, y_pred_svc_p)
```

```python
# This can also be used with polynomial.
# kernal is used to specify type of algo
# We can also mention different degrees

svc_p2 = SVC(kernel='poly', degree = 2)
svc_p2.fit(X_train, y_train)
y_pred_svc_p2 = svc_p2.predict(X_test)
accuracy_score(y_test, y_pred_svc_p2)
```

```python
# This is mainly used for numeric data

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
accuracy_score(y_test, y_pred_gnb)
```

```python
## This cannot be used with negative data
# Here standard scaling will not be applied
# This is mainly used for Text Data

from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
y_pred_mnb = mnb.predict(X_test)
accuracy_score(y_test, y_pred_mnb)
```

```python
# This is mainly used for Text Data

from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
y_pred_bnb = bnb.predict(X_test)
accuracy_score(y_test, y_pred_bnb)
```
